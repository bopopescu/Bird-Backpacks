```python
import serial
```

```python
serial_port = '/dev/cu.usbmodem14101';
baud_rate = 9600; 
write_to_file_path = "serialoutput.txt";
```

```python
output_file = open(write_to_file_path, "w+");
seri = serial.Serial(serial_port, baud_rate)
while True:
    ln = seri.readline();
    ln = ln.decode("utf-8") 
    print(ln);
    output_file.write(ln);
```
