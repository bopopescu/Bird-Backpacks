```python
import serial
import time
```

```python
serial = serial.Serial('/dev/cu.usbmodem14101', 9600)
##time.sleep(2)
```

```python
data =[]                       
for i in range(50):
    b = serial.readline()         
    string_n = b.decode()   
    string = string_n.rstrip() 
    flt = float(string)        
    print(flt)
    data.append(flt)           
    time.sleep(0.1)       

serial.close()
```

```python
for line in data:
    print(line)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.plot(data)
plt.xlabel('Time (sec)')
plt.ylabel('Photoresistor values')
plt.title('Photoresistor values over time')
plt.show()
```
